```python
def col_diff(cols1, cols2, yr1, yr2):
    ''' Takes two lists of column names and returns all differing columns.'''
    diff = []
    for c in cols2:
        if c not in cols1:
            diff.append(('in ' + str(yr2) + ', missing in ' + str(yr1), c))
    for c in cols1:
        if c not in cols2 and c not in diff:
            diff.append(('in ' + str(yr1) + ', missing in ' + str(yr2), c))
    return diff
```

```python
# Remove symbols and convert to numeric - add additional column for to and from
wrop = h1b['WAGE_RATE_OF_PAY'].map(str)
wrop_split = [x.split('-') for x in list(wrop)]
```

```python
wage_rate_of_pay_lower = []
wage_rate_of_pay_upper = []
```

```python
for x in wrop_split:
    if len(x) == 1:
        wage_rate_of_pay_lower.append(x[0])
        wage_rate_of_pay_upper.append(x[0])
    else:
        wage_rate_of_pay_lower.append(x[0])
        wage_rate_of_pay_upper.append(x[1])
```

```python
wage_strip_upper = [x.strip() for x in wage_rate_of_pay_upper]
wage_rate_of_pay_lower_num = [0 if x == 'nan' or x == '' else float(x) for x in wage_rate_of_pay_lower]
wage_rate_of_pay_upper_num = [0 if x == 'nan' or x == '' or x == 'N/A' else float(x) for x in wage_strip_upper]
```

```python
# add to h1b and drop original column
h1b.drop(['WAGE_RATE_OF_PAY'], axis = 1, inplace = True)
h1b['WAGE_RATE_OF_PAY_LOWER_BOUND'] = wage_rate_of_pay_lower_num
h1b['WAGE_RATE_OF_PAY_UPPER_BOUND'] = wage_rate_of_pay_upper_num
```

```python
# look for codes that don't fit the pattern: xx-xxxx
bad_codes = []
analyzed_codes = []
soc_codes = h1b['SOC_CODE'].map(str)
for index, code in enumerate(soc_codes):
    string = str(code)
    result1 = re.match(pattern1, string)
    result2 = re.match(pattern2, string)
    result3 = re.match(pattern3, string)
    result4 = re.match(pattern4, string)
    result5 = re.match(pattern5, string)
    result6 = re.match(pattern6, string)
    result7 = re.match(pattern7, string)
    if (result1 is None and result2 is None and result3 is None 
        and result4 is None and result5 is None 
        and result6 is None and result7 is None):
        bad_codes.append((index, code))
    elif result2 is not None:
        string.replace('.', '-')
    elif result3 is not None:
        string = string[0:3] + '-' + string[3:]
    elif result4 is not None:
        string = string[0:2] + '-' + string[2:4] + string[5:]
    elif result5 is not None:
        string = string[0:2] + '-' + string[2:4] + string[5:]
    elif result6 is not None:
        string.replace('--', '-')
    elif result7 is not None:
        string.strip()
    analyzed_codes.append(string)
```
